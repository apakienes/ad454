```r
library(data.table)
library(tidyverse) 
library(BBmisc) 
library(class) 
library(gmodels) 
library(plotly) 
options(warn=-1) 
library(fastDummies)
options(repr.matrix.max.rows=20, repr.matrix.max.cols=15) 

datapath <- "~/data_ad454"

```

```r
mybank <- read.csv(sprintf("%s/MyBank.csv", datapath))
mybank1 <- mybank %>% as.data.table
```

```r
mybank_train <-mybank1[1:(.N - 2000),]
mybank_test <- mybank1[.N - 1999:0,]
```

```r
mybank_test1 <- mybank_test %>% add_row(Age = 40, Experience = 10, Income = 84, Family = 2, CCAvg = 2,
                                        Education = 0, Mortgage = 0,Securities_Account = 0, CD_Account = 0,
                                        Online = 1, CreditCard = 1)
```

```r
mybank_test2 <- mybank_test1[,BBmisc::normalize(.SD, "range"), .SDcols = -c(1,5,10)]
mybank_train2 <- mybank_train[,BBmisc::normalize(.SD, "range"), .SDcols = -c(1,5,10)]
```

```r
mybank_train_labels <- mybank1[1:(.N - 2000)][[10]]
mybank_test_labels <- mybank1[.N - 1999:0][[10]]
```

```r
mybank_test_pred_n <- class::knn(train = mybank_train2,
                            test = mybank_test2,
                            cl = mybank_train_labels,
                            k = 3)
```
